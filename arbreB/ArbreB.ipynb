- Noeuds : m <= k <= 2m clés
- Racine : 1 <= k <= 2m clés

# Tests de suppression <a class="anchor" id="suppr"></a>

* Test 1 - Suppression au niveau d'une feuille avec une fusion avec un frère

```python
btree = ArbreB(2)
for i in range(20):
    btree.ajout(i)

dot = btree.visualiser_arbre()
dot
```

```python
# Fusion de la feuille [12, 13] avec son frere gauche [9, 10]
# Descente de la cle 11 du parent dans la feuille resultante
btree.suppression(13)
dot0 = btree.visualiser_arbre()
dot0
```

* Test 2 - Suppression au niveau d'un noeud interne avec emprunt de clé à un frère

```python
btree = ArbreB(2)
for i in range(21):
    btree.ajout(i)
dot = btree.visualiser_arbre()
dot
```

```python
btree.suppression(17)
dot4 = btree.visualiser_arbre()
dot4
```

* Test 3 - Suppression de la racine avec des fusions en remontée et diminution de la hauteur de l'arbre

```python
btree = ArbreB(2)
for i in range(17):
    btree.ajout(i)
dot = btree.visualiser_arbre()
dot
```

```python
# Suppression de la racine 8 -> Remplacement par la cle predecesseur 7 et suppression de 7
# Chaque noeud de la feuille [6, 7] a la racine possede le minimum de cles ainsi que son frere
# -> Fusion en remontee jusqu'a la racine et diminution de la hauteur :
# Fusion de [3, 4] et [6] -> le parent descend la cle 5 et devient [2]
# Fusion du parent [2] et de son frere [11, 14]
# -> La racine descend la cle 7 et devient vide -> Diminution de la hauteur de l'arbre
btree.suppression(8)
dot6 = btree.visualiser_arbre()
dot6
```
