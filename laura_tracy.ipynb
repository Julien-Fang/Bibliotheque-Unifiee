```python
from math import floor,log

class AVLnoeud:
    def __init__(self,cle):
        self.cle = cle
        self.nbNoeuds = 1
        self.parent = None
        self.gauche = None
        self.droite = None


class AVL:
    def __init__(self):
        self.racine = None

    def estAVLvide(self):
        return self.racine is None

    def AVLracine(self):
        return self.racine
    
    def AVLsousArbreDroit(self):
        if self.estAVLvide():
            return None
        else:
            return self.racine.droit

    def AVLsousArbreGauche(self):
        if self.estAVLvide():
            return None
        else:
            return self.racine.gauche

    def AVLhauteur(self):
        if self.estAVLvide():
            return 0
        if self.racine.gauche is None or self.racine.droite is None:
            return 1
        else:
            return floor(log(self.racine.nbNoeuds,2))
    
    def AVLrotationGauche(self):
        if self.estAVLvide():
            return

        racineDroite = self.AVLsousArbreDroit() # nouvelle racine
        droiteSousArbreGauche = racineDroite.AVLsousArbreGauche()
        
        racine = self.AVLracine() # ancienne racine
        racineDroite.gauche = racine  # racine devient fils gauche
        
        racine.droite = droiteSousArbreGauche # ancien fils gauche (du droit) devient fils droit de l'ancienne racine
        
        #update nbNoeuds
        racine.nbNoeuds = racine.nbNoeuds - racineDroite.nbNoeuds + droiteSousArbreGauche.nbNoeuds
        racineDroite.nbNoeuds = racineDroite.nbNoeuds - droiteSousArbreGauche.nbNoeuds + racine.nbNoeuds
        
        #update parent
        racine.parent = racineDroite
        racineDroite.parent = None
        
        self.racine = racineDroite
        

    def AVLrotationDroite(self):
        if self.estAVLvide():
            return

        racineGauche = self.AVLsousArbreGauche() # nouvelle racine
        gaucheSousArbreDroit = racineGauche.AVLsousArbreDroit()
        
        racine = self.AVLracine() # ancienne racine
        racineGauche.droite = racine  # racine devient fils droit
        
        racine.gauche = gaucheSousArbreDroit # ancien fils droit (du gauche) devient fils gauche de l'ancienne racine
        
        #update nbNoeuds
        racine.nbNoeuds = racine.nbNoeuds - racineGauche.nbNoeuds + gaucheSousArbreDroit.nbNoeuds
        racineGauche.nbNoeuds = racineGauche.nbNoeuds - gaucheSousArbreDroit.nbNoeuds + racine.nbNoeuds
        
        #update parent
        racine.parent = racineGauche
        racineGauche.parent = None
        
        self.racine = racineGauche

    def AVLrotationGaucheDroite(self):
        if self.estAVLvide():
            return
        self.AVLsousArbreGauche().AVLrotationGauche()
        self.AVLrotationDroite()

    def AVLrotationDroiteGauche(self):
        if self.estAVLvide():
            return
        self.AVLsousArbreDroit().AVLrotationDroite()
        self.AVLrotationGauche()


    def AVLfacteurEquilibre(self):
        if self.estAVLvide():
            return 0
        if self.AVLsousArbreGauche() is None:
            return -self.AVLhauteur()
        if self.AVLsousArbreDroit() is None:
            return self.AVLhauteur()
        return self.AVLsousArbreGauche().AVLhauteur() - self.AVLsousArbreDroit().AVLhauteur()
    
    def AVLequilibrage(self):
        if self.estAVLvide():
            return
        if self.AVLfacteurEquilibre() == 2:
            if self.AVLsousArbreGauche().AVLfacteurEquilibre() == 1:
                self.AVLrotationDroite()
            else:
                self.AVLrotationGaucheDroite()
        elif self.AVLfacteurEquilibre() == -2:
            if self.AVLsousArbreDroit().AVLfacteurEquilibre() == -1:
                self.AVLrotationGauche()
            else:
                self.AVLrotationDroiteGauche()
        else:
            return
        
        
    # def AVLequilibrage(self):
    #     if self.estAVLvide() or self.gauche is None or self.droite is None: # ok difference hauteur = 1
    #         return self
    #     hauteurGauche = self.gauche.hauteurAVL()
    #     hauteurDroite = self.droite.hauteurAVL
    #     if 
        

    def AVLajout(self, cle):
        if self.estAVLvide():
            return AVL(cle)

        if cle == self.AVLracine():
            return self
        
        self.racine.nbNoeuds += 1
        if cle < self.AVLracine():
            # rac = self.AVLracine()
            # nouveau_gauche = self.AVLajout(cle,self.AVLsousARbreGauche())
            # droite = self.AVLsousARbreDroit()
            # avl = AVL
            # return self.AVLequilibrage(ArbreBinaire(rac,nouveau_gauche,droite))
            nouveau_gauche = self.AVLsousArbreGauche().AVLajout(cle)
            self.gauche = nouveau_gauche
            return self.AVLequilibrage()       
        else:
            nouveau_droite = self.AVLajout(cle,self.AVLsousArbreDroit())
            self.droite = nouveau_droite
            return self.AVLequilibrage()
        

    def AVLsuppression(self,cle):
        pass

    def AVLrecherche(self,cle):
        pass
```
